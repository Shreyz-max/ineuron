## Task:
    
Create a sql db from adult dataset and name it sqladb

```python
conn = sql.connect("sqladb.db")
adult = pd.read_csv('adult.data')
adult
```

```python
cur = conn.cursor()
adult.to_sql('adult', conn, if_exists='append', index = False)
```

1. Select 10 records from the adult sqladb

```python
cur.execute("select * from adult limit 10;")
result = cur.fetchall()
for row in result:
    print(row)
```

2. Show me the average hours per week of all men who are working in private sector

```python
st = "SELECT AVG(hours_per_week) FROM adult WHERE sex=' Male' and workclass =' Private'"
cur.execute(st)
result = cur.fetchall()
print(result)
```

3. Show me the frequency table for education, occupation and relationship, separately

```python
st = "SELECT education, COUNT(*) FROM adult GROUP BY education"
cur.execute(st)
result = cur.fetchall()
print('By Education: ')
for row in result:
    print(row)
    
print('\nBy Occupation: ')
st = "SELECT occupation, COUNT(*) FROM adult GROUP BY occupation"
cur.execute(st)
result = cur.fetchall()
for row in result:
    print(row)
    
print('\nBy Relationship: ')
st = "SELECT relationship, COUNT(*) FROM adult GROUP BY relationship"
cur.execute(st)
result = cur.fetchall()
for row in result:
    print(row)    
```

4. Are there any people who are married, working in private sector and having a masters
degree

```python
st = "SELECT * FROM adult WHERE marital_status=' Married-civ-spouse' and workclass=' Private' and education=' Masters'"
cur.execute(st)
result = cur.fetchall()
for row in result[:10]:
    print(row)    
```

5. What is the average, minimum and maximum age group for people working in different
sectors

```python
st = "SELECT workclass, AVG(age), MAX(age), MIN(age) FROM adult GROUP BY workclass"
cur.execute(st)
result = cur.fetchall()
for row in result:
    print('For sector{} average age is {}, maximum age is {}, minimum age is {}'.format(row[0], row[1], row[2], row[3])) 
```

6. Calculate age distribution by country

```python
st ='''select  native_country,
    case 
        when age between 18 and 25 then '18-25' 
        when age between 26 and 35 then '26-35'
        when age between 36 and 45 then '36-45'
        when age between 46 and 55 then '46-55'
        else '55+'
    end as ages, 
    count(*) 
from adult
group by  native_country, ages'''
cur.execute(st)
result = cur.fetchall()
for row in result[:10]:
    print(row)
```

7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and
'capital-loss'

```python
st = "select *, (capital_gain - capital_loss) as net_capital_gain from adult"
cur.execute(st)
result = cur.fetchall()
for row in result[:10]:
    print(row)
```

1. Create an sqlalchemy engine using a sample from the data set

```python
import sqlalchemy as db
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy.sql import select, func
from sqlalchemy.orm import Session
engine = db.create_engine('mysql://root:password@localhost/sqladb')
conn = engine.connect()
```

```python
metadata = MetaData()
adult = Table('adult', metadata,
        Column('age', Integer),
        Column('workclass', String),
        Column('fnlwgt', Integer),
        Column('education', String),
        Column('education_num', Integer),
        Column('marital_status', String),
        Column('occupation', String),
        Column('relationship', String),
        Column('race', String),
        Column('sex', String),
        Column('capital_gain', Integer),
        Column('capital_loss', Integer),
        Column('hours_per_week', Integer),
        Column('native_country', String),
        Column('salary', String)
    )
```
